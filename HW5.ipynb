```python
############ Homework 5 Monte Carlo Techniques ############
############ 1. Importance of Randomness ############
############ MonetCarlo_integration.ipynb Excerise 1 for different random number generators############

%matplotlib inline
import numpy as np
from matplotlib import pyplot

# Hit and miss Monte Carlo integration
ngroups = 16

I = np.zeros(ngroups)
N = np.zeros(ngroups)
E = np.zeros(ngroups)

n0 = 100

for i in range(ngroups):
    N[i] = n0
    
############ Insert Different random number generators here ############
    x = np.zeros(n0+1)
    y = np.zeros(n0+1)
    a = 5
    M = 11
    seed = 1.
    x[0] = seed
    y[0] = seed
    for k in range(1,n0+1):
        x[k] = (a*x[k-1])%M
    for k in range(1,n0+1):
        x[k] = x[k]/M
    for l in range(1,n0+1):
        y[l] = (a*y[l-1])%M
    for l in range(1,n0+1):
        y[l] = y[l]/M
    
    I[i] = 0.
    Nin = 0
    for j in range(n0+1):
        if(y[j] < np.sqrt(1-x[j]**2)):
            Nin += 1
            
    I[i] = 4.*float(Nin)/float(n0)
    E[i] = abs(I[i]-np.pi)
    print (n0,Nin,I[i],E[i])
    n0 *= 2
            
pyplot.plot(N,E,ls='-',c='red',lw=3,label='E');
pyplot.plot(N,1/np.sqrt(N),ls='-',c='blue',lw=3,label='1/sqrt(N)');
pyplot.xscale('log')
pyplot.yscale('log')
pyplot.title('LCG parameters a=5, c=0, seed=1')
pyplot.legend()
pyplot.show()
```
